```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.info()
```

```python
mean_ages = df.groupby('Pclass')['Age'].mean()
```

```python
age1 = mean_ages[1] 
age2 = mean_ages[2] 
age3 = mean_ages[3]
```

```python
def conver_age(row):
    if pd.isnull(row['Age']):
        if row['Pclass'] == 1:
            return age1
        elif row['Pclass'] == 2:
            return age2
        elif row['Pclass'] == 3:
            return age3
    else:
        return row['Age']
df['Age'] = df.apply(conver_age, axis=1)
```

```python
df.head()
```

```python

df['Sex'].value_counts()
```

```python
def conver_sex(value):
    if value == 'male':
        return 1
    else:
        return 0

df['Sex'] = df['Sex'].apply(conver_sex)
```

```python
df['Sex'].value_counts()
```

```python
df.info()
```

```python
df.drop(['Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1, inplace=True)
```

```python
df['Alone'] = df['SibSp'] + df['Parch']
```

```python
df.info()
```

```python
df['Alone'].value_counts()
```

```python
df['Alone'] = df['SibSp'] + df['Parch']
```

```python
def conver_alone(value):
    if value != 0:
        return 1
    else:
        return 0 

df['Alone'] = df['Alone'].apply(conver_alone)
```

```python
df['Alone'].value_counts()
```

```python
df.head()
```

```python
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
```

```python
df.drop('Name', axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
x = df.drop('Survived', axis=1)
```

```python
y = df['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)
```

```python
classifier.fit(x_train,y_train)
```

```python
prediction = classifier.predict(x_test)
```

```python
print('Percentage:  ', accuracy_score(y_test, prediction) * 100)
```

```python
print('Confusion matrix: ', confusion_matrix(y_test, prediction, labels=[0,1]))
```
